```python
#@title Cargando LLAMA3
import transformers
import torch

model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.float16,
        "quantization_config": {"load_in_4bit": True},
        "low_cpu_mem_usage": True,
    },
)
```

```python
from IPython.display import display
import numpy
import pandas as pd
import requests
```

```python
# Cargar el archivo CSV en un DataFrame
search_df = pd.read_csv('scopus.csv')

# Mostrar las primeras filas del DataFrame
## print(search_df)

# Mostramos usando la funcion display
display(search_df)
```

# **Resolución 3.1 labore una lista de python con 3 preguntas dirigidas al abstract.**

```python

lista_preguntas = ["Dame un resumen en español de 50 palabras del abstract que te proporciono",
                   "¿Cuál es la eficiencia de retención del material o del dispositivo después de realizar múltiples ciclos de carga y descarga?",
                   "Cuál material presenta el mayor valor de capacitancia y qué valor específico tiene? ",
                   "Qué materiales se ha utilizado para fabricar los electrodos?",
                   "¿Síntesis para la obtención del carbón activado?"]
```

# **Resolución 3.2 Disponga de LLAMA y realice estas consultas almacenando las respuestas en archivos de texto o en el mismo DataFrame de pandas (según le convenga)**
En este caso hacimos 05 preguntas

```python
def consultas(input_text,question):
  messages = [
    {"role"   : "system",
     "content": input_text},
    {"role"   : "user",
     "content": lista_preguntas[question]},
  ]

  prompt = pipeline.tokenizer.apply_chat_template(
          messages,
          tokenize=False,
          add_generation_prompt=True
  )

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      prompt,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )

  return outputs[0]["generated_text"][len(prompt):]
```

```python
# Crear las columnas para las preguntas dinámicamente
for j in range(1, 6):
    search_df[f'Pregunta{j:02}'] = None

# Iterar sobre las preguntas y los abstracts
# Son 12 articulos los que se van a revisar y son 5 preguntas que se harán
for i in range(12):
    input_text = search_df["Abstract"][i]
    for j in range(5):
        response = consultas(input_text, j)
        search_df.iloc[i, search_df.columns.get_loc(f'Pregunta{j+1:02}')] = response
```

```python
# Mostramos usando la funcion display
display(search_df)
```

# **Resolución 3.3 Asigne un indicador numérico (como por ejemplo un puntaje) para luego filtrar su dataset.**
En este caso tenemos 2 columnas como indicadores numéricos, dependiendo del enfoque de cada una

```python

lista_puntaje = ["Eres un especialista en supercapacitores: Dame la capacitancia en medida Faradios por gramo (F/g). Recuerda, solo dame el numero, no me des más texto adicional.",
                   "Eres un especialista en supercapacitores:  Asigna una puntuación de 1 si el material utilizado ecomo electrodo es apto y eficiente para utilizarse en supercapacitores, y una puntuación de 0 si el material no es apto o no es eficiente para este uso, si no tienes la información suficiente para responder devuelve -1. Recuerda, solo dame el numero, no me des más texto adicional."]
```

```python
def puntaje(input_text,question):
  messages = [
    {"role"   : "system",
     "content": input_text},
    {"role"   : "user",
     "content": lista_puntaje[question]},
  ]

  prompt = pipeline.tokenizer.apply_chat_template(
          messages,
          tokenize=False,
          add_generation_prompt=True
  )

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      prompt,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )
  return outputs[0]["generated_text"][len(prompt):]
```

```python
# Crear las columnas para las preguntas dinámicamente
for j in range(1, 3):
    search_df[f'Puntaje{j:02}'] = None

# Iterar sobre las preguntas y los abstracts
# Son 12 articulos los que se van a revisar y son 5 preguntas que se harán
for i in range(12):
    input_text = search_df["Abstract"][i]
    for j in range(2):
        response = puntaje(input_text, j)
        search_df.iloc[i, search_df.columns.get_loc(f'Puntaje{j+1:02}')] = response
```

```python
# Mostramos usando la funcion display
display(search_df["Puntaje01"])
# Puntaje01 es la capacitancia, es el puntaje numerico que vamos a usar para filtrar
```
