# **Resolución 2.2 Usando matplotlib y/o seaborn grafique un diagrama de barras con los tipos de publicación versus la cantidad.**
### Vamos a hacer graficos acorde al tipo de publicación: Artículos, conference papers, books, etc
Aqui debe notarse la siferencia, en el csv que obtenemos, vamos a basarnos en la columna "Document Type", por lo que no podremos usar el subtype_description de los ejemplos

```python
search_df["Document Type"].unique()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Contar las ocurrencias de cada tipo de publicación y ordenar los datos por el índice
subtype_description = search_df["Document Type"].value_counts()

# Crear el gráfico de barras con coloración por tipo de publicación
ax = sns.barplot(x=subtype_description.index, y=subtype_description.values, palette='viridis')

# Agregar etiquetas y título
plt.ylabel('Cantidad')
plt.title('Tipo de publicación')

# Formatear el eje y con un decimal
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))

# Mostrar el gráfico
plt.show()
```
